## 자료구조와 배열

````python
# 학생 5명의 시험 점수를 입력받아 합계와 평균을 출력하기

print('학생 그룹 점수의 합계와 평균을 구합니다.')

score1 = int(input('1번 학생의 점수를 입력하세요: '))
score2 = int(input('2번 학생의 점수를 입력하세요: '))
score3 = int(input('3번 학생의 점수를 입력하세요: '))
score4 = int(input('4번 학생의 점수를 입력하세요: '))
score5 = int(input('5번 학생의 점수를 입력하세요: '))

total = 0
total += score1
total += score2
total += score3
total += score4
total += score5

print(f'학생 그룹 점수의 합계: {total}')
print(f'학생 그룹 점수의 평균: {total / 5}')
````

## Notes

##### 배열 

- 배열을 사용하면 따로 따로 흩어진 변수를 하나로 묶어서 사용할 수 있어 코드를 쉽고 효율적으로 작성할 수 있다.

- 하나의 값을 저장하는 변수가 아닌 묶음 단위로 값을 저장해야 하는 경우, 배열이라는 자료구조를 다루어야 하고, 배열에는 객체가 저장되며, 배열에 저장된 객체 하나를 원소라고 한다. 또한 각 원소는 0,1,2 순으로 인덱스를 부여받는다.

- 파이썬에는 배열을 리스트와 튜플로 구현할 수 있다. 리스트는 원소를 변경할 수 있는 뮤터블 list형 객체이다. 리스트는 연산자 [] 안에 쉼표로 구분하여 원소를 표기하고 생성한다. 리스트나 튜플의 원소에 접근할 때는 *인덱스*를 사용해 접근 할 수 있다. *슬라이싱*의 경우, 리스트 또는 튜플의 원소 일부를 연속해서 또는 일정한 간격으로 꺼내 새로운 리스트나 튜플을 만드는 것이다.

```
s[:] - 리스트 s의 원소를 모두 출력
s[:n] - 리스트 s의 원소 중 맨 앞부터 n개까지 출력
s[i:] - 리스트 s의 원소중 s[i] 부터 끝까지 출력
```

##### 뮤터블(mutable, 값을 변경할 수 있음)과 이뮤터블(immutable, 값을 변경할 수 없음)

- 변수에 값이 저장되어 있어도 다른 값을 대입할 수 있다. 값이 변경되면 값을 복사하는 것이 아니라 값을 참조하는 객체의 식별번호가 변경된다. 

```
n = 5 # n에 int형 정숫값 5를 대입
id(n) # id() 함수로 객체의 식별 번호를 출력
>>> 1213012412
n = 'ABC'
>>> 1213012444 # 함수의 식별 번호 변경
```

- 뮤터블 자료형 : 리스트, 딕셔너리, 집합 등이 있으며 값을 변경할 수 있다.
- 이뮤터블 자료형 : 수, 문자열, 튜플 등이 있으며 값을 변경할 수 없다.

##### 파이썬의 대입 과정
```
좌변에 변수 이름이 처음 나온 경우, 그 변수에 맞는 자료형으로 자동 선언해 준다.
대입식은 값 자체가 아니라 참조하는 객체의 식별 번호를 대입한다.
여러 변수에 여러 값을 한꺼번에 대입할 수 있다.
```

##### 자료구조

- 자료구조는 논리적인 관계로 이루어진 데이터 구성을 말한다. 데이터 단위와 데이터 자체 사이의 물리적 또는 논리적인 관계를 의미한다.

- 자료구조를 앎으로써, 컴퓨터에서 처리해야 하는 많은 데이터를 모아 효율적으로 관리하고 구조화 할 수 있다.

##### 등가성 vs. 동일성

- 파이썬에서는 값을 비교할 때 등가성(equality)과 동일성(identity)를 사용한다. 등가성 비교는 ==을, 동일성 비교는 is를 사용한다. 등가성 비교는 좌변과 우변의 값은지 비교하고, 동일성 비교는 값은 물론 객체의 식별번호까지 같은지 비교한다.
