````python
# 세 정수의 최댓값 구하기

def max3(a, b, c):
    '''a, b, c의 최댓값을 구하여 반환'''
    maximum = a
    if b > maximum: maximum = b
    if c > maximum: maximum = c
    return maximum

print(f'max3(3, 2, 1) = {max3(3, 2, 1)}')
print(f'max3(3, 1, 2) = {max3(3, 1, 2)}')
print(f'max3(2, 3, 1) = {max3(2, 3, 1)}')
print(f'max3(2, 1, 3) = {max3(2, 1, 3)}')
print(f'max3(1, 3, 2) = {max3(1, 3, 2)}')
````

````python
# 세 정수를 입력받아 중앙값 구하기

def med3(a,b,c):
    '''a, b, c의 중앙값을 구하여 반환'''
    if a >= b:
        if b >= c:
            return b
        elif a <= c:
            return a
        else:
            return c
    elif a > c:
        return a
    elif b > c:
        return c
    else:
        return b

print('세 정수의 중앙값을 구합니다.')
a = int(input('정수 a의 값을 입력하세요: ')) 
b = int(input('정수 b의 값을 입력하세요: '))
c = int(input('정수 c의 값을 입력하세요: '))

print(f'중앙값은 {med3(a, b, c)}입니다.')
````

````python
# 1부터 n까지 정수의 합 구하기 1(while 문)

print('1부터 n까지 정수의 합을 구합니다.')
n = int(input('n값을 입력하세요.: '))

sum = 0
i = 1

while i <= n:
    sum += i
    i += 1

print(f'1부터 {n}까지 정수의 합은 {sum}입니다.')
````

## Notes

##### 순차구조와 선택구조

- 순차구조 : 한 문장씩 순서대로 처리되는 구조

- 선택구조 : 조건식으로 평가한 결과에 따라 프로그램의 실행흐름이 변경되는 구조

- 알고리즘의 흐름은 위에서 아래 방향이며, 조건식이 있을 경우, 조건식의 평가 결과에 따라 '예'나 '아니오' 중 하나를 따라가고, 이를 양 갈래 선택이라고 한다.

##### input() 함수와 형 변환

- input() 함수는 키보드로 문자열을 입력받아 반환한다. 실행결광와 같이 input() 함수를 호출하면 입력받은 문자열형을 얻고 그 문자열이 변수에 대입된다.

- 그런데 만약 저장해야 할 데이터의 자료형이 문자열이 아닌 10진수 정수형일 경우, 변환해주어야 하고, 이럴 경우 int() 함수를 사용하여 형 변환을 할 수 있다.

- float은 부동 소수점 방식을 사용한다. 부동 소수점은 컴퓨터에서 실수를 근삿값으로 표현할 때 사용한다. 부동 소수점 방식은 실수를 기수부분과 지수부분으로 나누어 표현하는 것을 말한다. 이때, 기수 부분은 유효 숫자를 나타내고 지수 부분은 소수점의 위치를 나타낸다.

##### 복합문의 구조

- if 문이나 while 문 등 복합문의 첫 부분은 if나 while과 같은 키워드로 시작하여 콜론으로 끝난다. 이 부분을 헤더라고 한다. 헤더의 마지막 콜론은 바로 뒤에 스위트가 이어진다는 의미이다. 스위트는 헤더와 한 세트로 따라다니는 실행문을 의미한다.

##### 함수의 반환값과 함수 호출식 평가하기

- 함수 내부에서 처리한 값을 반환할 때에는 return 문을 사용하면 된다. 실제로 함수가 반환하는 값을 얻으려면 함수를 호출해야 하며, 이를 '함수 호출식을 평가해야 함수가 반환한 값을 얻을 수 있다.'고 한다. 

##### 복합문을 작성할 때 지켜야 할 규칙

- 복합문의 스위트는 반드시 행마다 같은 수준으로 들여쓰기를 해야한다. 들여쓰기는 공백을 최소 1개 이상 사용해야 하며, PEP 8에서는 공백 4개를 들여쓰기로 사용할 것을 권장한다.
```
if a < b:
    min 2 = a # 공백 4개로 들여쓰기
    max 2 = b # 이 문장도 공백 4개로 들였긔를 해야 if a < b:에 종속되어 실행된다.
```

- 만약 스위트가 단순문이면 헤더와 같은 행에 둘 수 있고, 또한 단순문이 2개 이상이면 각각의 단순문을 세미콜론으로 구분하여 헤더와 같은 행에 둘 수 있다.
```
if a < b : min2 = a # 단순문 1개
if a < b : min2 = a; max2 = b # 단순문 2개
```

##### 조건문과 분기

- 분기(branching)는 프로그램의 실행 흐름을 다른곳으로 변경하는 명령을 의미한다. 프로그램이 실행된다면 분기 중 어느 하나만 그리고 어느 하나는 무조건 실행된다.

##### 연산자(operator)와 피연산자(operand)

- 프로그래밍 언어에서는 +나 -등의 기호를 산술 연산자, 연산 대상을 피연산자라고 한다. 예를 들어, 대소 관계를 판단하는 식 a > b에 연산자는 >이고, 피 연산자는 a와 b이다.
```
단항 연산자(unary operator): 피연산자 1개 ex. -a
이항 연산자(binary operator): 피연산자 2개 ex. a < b
삼항 연산자(ternary operator): 피연산자 3개 ex. a if b else c
```

- 조건 연산자인 if~else 문은 파이썬의 유일한 삼항 연산자이다. 조건식을 평가한 값이 참인지, 거짓인지에 따라 값을 출력한다.

##### 순서도

- 순서도는 문제를 정의, 분석하고 해결하는 방법을 그림으로 표현한다.
```
<프로그램 순서도의 기호>
- 실제로 실행할 연산을 나타내는 기호
- 제어 흐름을 나타내는 선 기호
- 프로그램 순서도를 이해하거나 작성하는데 편리한 특수 기호
```

- 데이터는 기억 장치를 지정하지 않은 데이터 자체를 나타낸다.

- 처리는 여러 종류의 처리 기능을 나타낸다. 즉, 정보의 값/형/위치를 바꾸도록 정의한 연산이나 연산 집합의 실행, 또는 연속하는 몇 가지 흐름 가운데 하나의 방향을 결정하는 연산이나 연산 집합의 실행을 나타낸다.

- 미리 정의한 처리는 서브루틴이나 모듈 등 다른 곳에서 이미 정의한 하나 이상의 연산 또는 명령으로 이루어진 처리를 나타낸다.

- 판단은 하나의 입구와 하나 이상을 선택하는 출구가 있고, 판단 기호 안에 정의한 조건을 평가하여 하나의 출구를 선택하는 판단 기능을 나타낸다.

- 루프 범위는 두 부분으로 구성되어 루프의 시작과 종료를 나타낸다.

- 선은 제어의 흐름을 의미한다.

- 단말은 외부 환경으로 나가거나 외부환경에서 들어오는 것을 나타내는 것으로, 주로 프로그램 흐름의 시작과 종료를 나타낸다.

##### while문 반복 구조

- 어떤 조건이 성립하는 동안 반복해서 처리(프로그램 명령문 또는 명렁어의 집합)하는 것을 반복 구조라 하고 일반적으로 루프라고 한다. 이때, while문은 실행하기 전에 반복을 계속할 것인지를 판단하는데 이런 구조를 *사전 판단 반복 구조*라고 한다. 그리고 반복 대상이 반복되는 명령문을 루프 본문이라고 한다.

- 반복을 제어할 때, 사용하는 i를 카운터용 변수라고 한다.

##### for문 반복구조

- 변수가 하나만 있을 때는 while 문보다 for 문을 사용하는 것이 좋다.

##### range() 함수로 이터러블 객체(반복할 수 있는 객체) 생성하기

- range() 함수는 이터러블 객체를 생성한다.
```
range(n) - 0 이상 n 미만인 수를 차례로 나열하는 수열
range(a, b) - a 이상 b 미만인 수를 차례로 나열하는 수열
range(a, b, step) - a 이상 b 미만인 수를 step 간격으로 나열하는 수열
```

##### 두 값 교환하기 알고리즘

- a와 b를 교환할 때 사용한 단일 대입문 a, b = b, a는 이 과정으로 수행된다.
```
1. 우변의 b,a에 의해 두 값을 압축한 튜플 (b,a)가 생성된다.
2. 대입할 때 튜플 (b,a)를 다시 풀어 b, a로 만든 다음 각각 a와 b에 대입한다.
```

##### 두 값 교환하기 알고리즘_2

- 단일 대입문으로 a와 b의 값을 교환하는 방법 외에 임시용 변수 t를 이용하여 a와 b의 값을 교환할 수도 있습니다.
```
1. a값을 t에 저장한다.
2. b값을 a에 대입한다.
3. t에 저장한 처음 a값을 b에 대입한다.
```


##### 무한 루프와 break 문

- while 문의 조건식에 True가 사용된 것은 의도적으로 while문이 무한 반복되도록 만든 것이며 이를 무한 루프(infinite loop)라고 한다.

- 반복문 안에서 break 문을 실행하면 반복문을 종료할 수 있다.

##### 파이썬의 변수

- 파이썬에서는 데이터, 함수, 클래스, 모듈, 패키지 등을 모두 객체로 취급한다. 객체는 자료형을 가지며 메모리를 차지한다. 파있너의 이런 특징 때문에 파이썬의 변수는 값을 갖지 않는다. 예를 들어 x = 17은 x가 17이라는 값을 갖고 있다고 말할 수 없다. 

- 변수는 객체를 참조하는 객체에 연결된 이름에 불과하다. 모든 객체는 메모리를 차지하고, 자료형 뿐만 아니라 식별 번호를 지닌다.

##### 지역변수와 전역변수

- 전역 변수는 프로그램 전체에서 사용할 수 있는 변수이고, 지역 변수는 함수 내부에서만 사용할 수 있는 변수이다.
